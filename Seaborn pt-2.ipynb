# Matrix Plot

```python
import seaborn as sns
import matplotlib as plt
```

```python
tips=sns.load_dataset("tips")
flights=sns.load_dataset("flights")
```

```python
tips.head(3)
```

```python
flights.head(3)
```

```python
tc=tips.corr()
tc
```

```python
sns.heatmap(tc,annot=True,cmap="coolwarm")
```

```python
ps=flights.pivot_table(index="year",columns="month",values="passengers")
ps
```

```python
sns.heatmap(ps,cmap="coolwarm")
```

```python
sns.clustermap(ps,cmap="coolwarm")
```

# Grids

```python
iris=sns.load_dataset("iris")
iris.head(4)
```

```python
sns.pairplot(iris)
```

```python
g=sns.PairGrid(iris)                            # to have more control over our graphs

g.map_diag(sns.distplot)
g.map_lower(sns.scatterplot)
g.map_upper(sns.kdeplot)


```

```python
h=sns.FacetGrid(data=tips,row='smoker',col="time")
# h.map(sns.distplot,'total_bill')                  #takes 1 argument
h.map(sns.scatterplot,'total_bill',"tip")           #takes 2 argumengts
```

# Regression Plot

```python
sns.lmplot(x="total_bill",y="tip",data=tips,row="time",markers=['o','v'],hue="sex")
```
