```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,7,11)
x
```

```python
y=x**2
y
```

Functional method:::::::::

```python
plt.plot(x,y,'g*')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Simple graph")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.title("Plot 1")

plt.subplot(1,2,2)
plt.plot(y,x,'g')
plt.title("Plot 2")
```

OOPS Method::::::::

```python
fig=plt.figure()

add=fig.add_axes([0.1,0.1,0.8,0.8])

add.plot(x,y,'y')
add.set_xlabel("X")
add.set_title("Title")
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.3,0.17,0.57,0.3])
 
axes1.plot(x,y,'r')
axes2.plot(y,x,'g')
```

```python
axis,fig=plt.subplots(nrows=1,ncols=2) # this format compulsary

fig[0].plot(x,y,'r')
fig[1].plot(x,y,'y')

plt.tight_layout()                    # propper arrengenment
```

# Figure Size and DPI

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(8,4))
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig.savefig("Graph.png")
```

```python
bird=plt.figure()
fly=bird.add_axes([0.1,0.1,0.8,0.8])        #    list pass karni hai

fly.plot(x,x**2,label="squared")            #label use kia hai legend ke liye
fly.plot(x,x**3,label="cubed")

fly.legend()
```

```python
sal=plt.figure()
plate=sal.add_axes([0,0,1,1])

plate.plot(x,y,"purple",lw=4,alpha=0.7,ls="--",marker="o",markerfacecolor="blue")     #linewidht   #transpiracy    #linestyle
```

Xlimit and y lilmit

```python
fig,snow=plt.subplots(nrows=1,ncols=2)

snow[0].plot(x,y,'red',ls="--")

snow[0].set_xlim([0,4])
snow[0].set_ylim([0,20])


snow[1].plot(y,x,"orange",ls=':',lw=4)
```
